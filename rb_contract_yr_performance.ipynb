```python
ngs = pd.read_csv("./rookie_contract_advanced_stats.csv")
ngs['pay_increase'] = ngs.apply(lambda r: ((r.fc - r.rc)/r.rc) * 100, axis = 1)
ngs['eff_total'] = ngs.apply(lambda r: ((r.eff2019 + r.eff2018 + r.eff2017)), axis = 1)
ngs['ryoe_total'] = ngs.apply(lambda r: ((r.ryoe2019 + r.ryoe2018)), axis = 1)
ngs['ryoe_total'] = ngs.apply(lambda r: 1 if ((r.ryoe2019 - r.ryoe2018 > 0)) else 0, axis = 1)
ngs
```

```python
ngs['avg_pay'] = ngs.apply(lambda r: 1 if ((r.pay_increase > 436.955479)) else 0, axis = 1)
ngs['cyeff'] = ngs.apply(lambda r: ((r.eff2019 - r.eff2018)), axis = 1)
ngs['gbc'] = ngs.apply(lambda r: 2 if (r.pay_increase < 436.955479 and r.eff2020 > 2.323043) else (1 if r.pay_increase > 436.955479 and r.eff2020 > 2.323043 else (0)), axis = 1)
ngs['yd2_yd3'] = ngs.apply(lambda r: (r.Yds2 - r.Yds3)/r.Yds3 if (r.Yds2  > r.Yds3) else ((1+(r.Yds2 - r.Yds3)/(r.Yds3)) if (r.Yds2 + 320 > r.Yds3) else ((r.Yds2 - r.Yds3)/(r.Yds3))), axis=1)
ngs['model_score'] = ngs.apply(lambda r: (((((2.608261*r.ryoe2018 * (r.eff_total - 7.524))) * ((r.yd2_yd3)) * (5*r.Yds2)))*(((r.fc-r.rc)/r.rc)/(436.955479*1000))), axis=1)
ngs
```

```python
ngs.groupby('model_score')[['Player','model_score']].max()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X = np.array(ngs.iloc[:,[14, 5, 6, 7, 8, 17, 19]])
y = np.array(ngs.iloc[:,[10]])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5)
lrm = LinearRegression().fit(X_train,y_train)
y_pred = lrm.predict(X_test)
print(X_test)
print(y_pred)
```
