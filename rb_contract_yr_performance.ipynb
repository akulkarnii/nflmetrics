```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#draft_class_2015 = ['Todd Gurley', 'Melvin Gordon', 'T.J. Yeldon', 'Ameer Abdullah', 'Tevin Coleman', 'Duke Johnson', 'David Johnson', 'Javorious Allen', 'Mike Davis', 'Jay Ajayi']
draft_2016_signed_2020 = ['Derrick Henry', 'Jordan Howard', 'Kenyan Drake', 'Tyler Ervin', 'Devontae Booker', 'DeAndre Washington']
draft_2017_signed_2020 = ['Alvin Kamara', 'Christian McCaffrey', 'Dalvin Cook', 'Joe Mixon', 'Tarik Cohen', 'Kareem Hunt']
draft_2017_signed_2021 = ['Aaron Jones', 'Christopher Carson', 'Jamaal Williams', 'Samaje Perine', 'Marlon Mack', 'James Conner']
draft_2018_signed_2021 = ['Nick Chubb', 'Nyheim Hines', 'Kerryon Johnson', 'Ito Smith', 'Kalen Ballage']
#signed year is first year new contract comes into play
player_stats = pd.read_csv("./year_before_contract.csv")
player_stats
```

```python
player_stats['fD%D'] = player_stats.apply(lambda r: ((r.fD2 - r.fD1)/r.fD1) * 100, axis = 1)
player_stats['y/a%D'] = player_stats.apply(lambda r: ((r.YA2 - r.YA1)/r.YA1) * 100, axis = 1)
player_stats['s%D'] = player_stats.apply(lambda r: ((r.fc - r.rc)/r.rc) * 100, axis = 1)
player_stats
```

```python
plt.scatter(np.array(player_stats.iloc[:,[12]]), np.array(player_stats.iloc[:,[25]]))
```

```python
from sklearn.cluster import KMeans
nnX = np.hstack((np.array(player_stats.iloc[:,[12]]), np.array(player_stats.iloc[:,[25]])))
kmeans = KMeans().fit(nnX)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
X = np.array(player_stats.iloc[:,[5,6,7,8,9,10,11,12,13]])

y = np.array(player_stats.iloc[:,[28]])
y2 = np.array(player_stats.iloc[:,[24]])
y3 = np.array(player_stats.iloc[:,[27]])
y4 = np.array(player_stats.iloc[:,[26]])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y2, test_size=0.40)
mlp = MLPRegressor(solver = 'lbfgs', activation = 'relu', max_iter=1000, hidden_layer_sizes=(1, 2)).fit(X_train,np.ravel(y_train))
mlp.score(X_test, y_test)
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression().fit(X_train,y_train)
print(lin.score(X_test, y_test))
y_pred_lin = lin.predict(X_test)
```
