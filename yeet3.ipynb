```python
import pandas as pd
centers_data = pd.read_csv("./centers.csv")
tackles_data = pd.read_csv("./tackles.csv")
guards_data = pd.read_csv("./guards.csv")
oline_data_2016 = pd.read_csv("./general_oline_2016.csv")
rb_data_2016 = pd.read_csv("./individual_rb_2016.csv")
nfl_data_2017 = pd.read_csv("./2017_nfl_rb_stats.csv")
nfl_data_2018 = pd.read_csv("./2018_nfl_rb_stats.csv")
nfl_data_2019 = pd.read_csv("./2019_nfl_rb_stats.csv")
print(centers_data.head(), "\n\n", tackles_data.head(), "\n\n", guards_data.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X = np.array(players_that_matter_2019.iloc[:,[2,3,4,5,6,8]])
y = np.array(nfl_data3.iloc[:,5]).reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
model = LinearRegression().fit(X_train,y_train)
print("score: ", model.score(X_test,y_test))
print("theta0: ", model.intercept_)
print("theta: ", model.coef_)
```
